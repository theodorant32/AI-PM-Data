# Exploratory Data Analysis

Analyzing weather patterns and relationships in the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned_weather.csv')
df['lastupdated'] = pd.to_datetime(df['lastupdated'])
print(f'Shape: {df.shape}')
```

```python
df.describe()
```

## Temperature Over Time

```python
# Sample to daily averages for cleaner visualization
daily_temp = df.groupby(df['lastupdated'].dt.date)['temperature'].mean()
plt.figure(figsize=(14, 6))
plt.plot(daily_temp.index, daily_temp.values, linewidth=1.5, color='red')
plt.title('Average Daily Temperature Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/plots/temperature_over_time.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Precipitation Over Time

```python
# Sample to daily averages for cleaner visualization
daily_precip = df.groupby(df['lastupdated'].dt.date)['precipitation'].mean()
plt.figure(figsize=(14, 6))
plt.plot(daily_precip.index, daily_precip.values, linewidth=1, color='blue', alpha=0.8)
plt.title('Average Daily Precipitation Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Precipitation (mm)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/plots/precipitation_over_time.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Distribution Plots

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df['temperature'], kde=True, ax=axes[0], color='red', bins=40)
axes[0].set_title('Temperature Distribution', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Temperature (°C)', fontsize=11)
axes[0].set_ylabel('Count', fontsize=11)

sns.histplot(df['humidity'], kde=True, ax=axes[1], color='steelblue', bins=40)
axes[1].set_title('Humidity Distribution', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Humidity (%)', fontsize=11)
axes[1].set_ylabel('Count', fontsize=11)

sns.histplot(df['precipitation'], kde=True, ax=axes[2], color='green', bins=40)
axes[2].set_title('Precipitation Distribution', fontsize=13, fontweight='bold')
axes[2].set_xlabel('Precipitation (mm)', fontsize=11)
axes[2].set_ylabel('Count', fontsize=11)

plt.tight_layout()
plt.savefig('../outputs/plots/distributions.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Correlation Heatmap

```python
# Select key variables only for readable heatmap
key_vars = ['temperature', 'humidity', 'precipitation', 'wind_kph', 'pressure_mb', 
            'cloud', 'uv_index', 'visibility_km', 'feels_like_celsius']
corr_matrix = df[key_vars].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, 
            linewidths=0.5, linecolor='white', annot_kws={'size': 10})
plt.title('Correlation Matrix of Key Weather Variables', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/plots/correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Key Insights

1. **Temperature patterns**: The time series shows clear seasonal variation with peaks and troughs corresponding to summer and winter periods.

2. **Precipitation**: Highly variable with sporadic spikes, typical of weather data where rain events are intermittent.

3. **Distributions**: Temperature shows a roughly normal distribution centered around the mean. Humidity is skewed toward higher values.

4. **Correlations**: Humidity and precipitation show positive correlation, which makes sense meteorologically. Temperature correlations with other variables will inform feature selection for modeling.
