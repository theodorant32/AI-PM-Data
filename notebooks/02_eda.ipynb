# Exploratory Data Analysis

Analyzing weather patterns and relationships in the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned_weather.csv')
df['lastupdated'] = pd.to_datetime(df['lastupdated'])
print(f'Shape: {df.shape}')
```

```python
df.describe()
```

## Temperature Over Time

```python
plt.figure(figsize=(12, 5))
plt.plot(df['lastupdated'], df['temperature'], linewidth=0.5, alpha=0.7)
plt.title('Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/plots/temperature_over_time.png', dpi=150)
plt.show()
```

## Precipitation Over Time

```python
plt.figure(figsize=(12, 5))
plt.plot(df['lastupdated'], df['precipitation'], linewidth=0.5, alpha=0.7, color='blue')
plt.title('Precipitation Over Time')
plt.xlabel('Date')
plt.ylabel('Precipitation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/plots/precipitation_over_time.png', dpi=150)
plt.show()
```

## Distribution Plots

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

sns.histplot(df['temperature'], kde=True, ax=axes[0], color='red')
axes[0].set_title('Temperature Distribution')
axes[0].set_xlabel('Temperature')

sns.histplot(df['humidity'], kde=True, ax=axes[1], color='blue')
axes[1].set_title('Humidity Distribution')
axes[1].set_xlabel('Humidity')

sns.histplot(df['precipitation'], kde=True, ax=axes[2], color='green')
axes[2].set_title('Precipitation Distribution')
axes[2].set_xlabel('Precipitation')

plt.tight_layout()
plt.savefig('../outputs/plots/distributions.png', dpi=150)
plt.show()
```

## Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10, 8))
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Weather Variables')
plt.tight_layout()
plt.savefig('../outputs/plots/correlation_heatmap.png', dpi=150)
plt.show()
```

## Key Insights

1. **Temperature patterns**: The time series shows clear seasonal variation with peaks and troughs corresponding to summer and winter periods.

2. **Precipitation**: Highly variable with sporadic spikes, typical of weather data where rain events are intermittent.

3. **Distributions**: Temperature shows a roughly normal distribution centered around the mean. Humidity is skewed toward higher values.

4. **Correlations**: Humidity and precipitation show positive correlation, which makes sense meteorologically. Temperature correlations with other variables will inform feature selection for modeling.
